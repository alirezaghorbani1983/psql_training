```python
%%sql  

CREATE TABLE users (
    id INT PRIMARY KEY,
    username VARCHAR(100)
);

INSERT INTO users (id, username) VALUES
(1, 'Reyna Marvin'),
(2, 'Micah Cremin'),
(3, 'Alfredo66'),
(4, 'Gerard_Mitchell42'),
(5, 'Frederique Donnelly');
```

```python
%%sql  
CREATE TABLE photos (
    id INT PRIMARY KEY,
    url VARCHAR(255),
    user_id INT,
    FOREIGN KEY (user_id) REFERENCES users(id)
);

INSERT INTO photos (id, url, user_id) VALUES
(1, 'https://santina.net', 1),
(2, 'https://alayna.net', 2),
(3, 'https://kailyn.name', 3);
```

```python
%%sql

select url,username
from photos
join users on users.id = photos.user_id;
```
