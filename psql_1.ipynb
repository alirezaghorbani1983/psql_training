## Using ON DELETE CASCADE

```python
%%sql
CREATE TABLE photos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    url VARCHAR(200),
    user_id INTEGER REFERENCES users(id) ON DELETE CASCADE
);

INSERT INTO photos (id, url, user_id) VALUES
(1, 'http://img1.jpg', 4),
(2, 'http://img2.jpg', 1),
(3, 'http://img3.jpg', 1),
(4, 'http://img4.jpg', 1),
(5, 'http://img5.jpg', 2),
(6, 'http://img6.jpg', 3),
(7, 'http://img6.jpg', 4);
```

```python
%%sql

delete from users 
where id = 1;
```

```python
%%sql
select * from photos;
```

## Using ON DELETE SET NULL

```python
%%sql
DROP TABLE IF EXISTS photos;
```

```python
%%sql
CREATE TABLE photos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    url VARCHAR(200),
    user_id INTEGER REFERENCES users(id) ON DELETE SET NULL
);

INSERT INTO photos (url, user_id) VALUES
('http://img1.jpg', 4),
('http://img5.jpg', 2),
('http://img6.jpg', 3),
('http://img6.jpg', 4);
```

```python
%%sql
delete from users
where id = 4;
```
